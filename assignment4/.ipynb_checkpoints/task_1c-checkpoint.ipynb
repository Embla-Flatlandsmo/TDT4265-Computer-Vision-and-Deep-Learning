```python
import matplotlib.pyplot as plt

xticks = [0.1*i for i in range(11)]

precision_1=[1.0, 1.0, 1.0, 0.5, 0.20]
recall_1 = [0.05, 0.1, 0.5, 0.7, 1.0]

precision_2 = [1.0, 0.8, 0.6, 0.5, 0.2]
recall_2 = [0.3, 0.4, 0.5, 0.7, 1.0]

plt.title("Class 1")
# plt.step(recall_1, precision_1, where='post')
plt.plot(recall_1, precision_1, '--')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.xlim([0,1])
plt.xticks(xticks)
plt.grid()
plt.show()

plt.title("Class 2")
# plt.plot(recall_2, precision_2)
# plt.step(recall_2, precision_2)
plt.plot(recall_2, precision_2, '--')
plt.grid(visible=True)
plt.xticks(xticks)
plt.xlim([0,1])
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.show()
```
